```python
X=df_train
y=df_train['NObeyesdad']
X.drop(['NObeyesdad'],axis=1,inplace=True)
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
y_test.unique()
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier

# Assuming X_train, y_train, X_test, y_test are defined
model = RandomForestClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)

model.predict(df_test)


accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, predictions))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators= [int(i) for i in np.linspace(100,2000,10)]
criterion=['gini', 'entropy', 'log_loss']
max_depth= [int(i) for i in np.linspace(100,5000,10)]
min_samples_split=[int(i) for i in np.linspace(2,20,19)]
min_samples_leaf=[int(i) for i in np.linspace(2,20,19)]
max_features=['sqrt', 'log2', None]
min_weight_fraction_leaf=[float(i) for i in np.linspace(0,0.5,6)]
```

```python
random_grid={'n_estimators':n_estimators,'criterion':criterion,'max_depth':max_depth,'min_samples_split':min_samples_split,
             'min_samples_leaf':min_samples_leaf,'max_features':max_features,'min_weight_fraction_leaf':min_weight_fraction_leaf}
rf=RandomForestClassifier()
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,n_iter=100,n_jobs=-1,cv=3,verbose=2,random_state=24)
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_estimator_
```
