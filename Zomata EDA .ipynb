```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
final_df.head()
```

```python
country_names=final_df.Country.value_counts().index #for country column
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%') #shows top 3 distribution of the column , to find the percentage we use the autopct="%1.2f%%"
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(15,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)#without any specified color,now in the next code cell we will see the colors which are specified in rating color
```

##Observation
1) 4.5 - 4.9 -->Excellent
2)4 - 4.4  --> Very Good
3)3.5 - 3.9 --> Good
4)2.5 - 3.4 --> Average
5)1.8 - 2.4 --> Poor


```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue="Rating color")#When we add hue it shows different colors but they need to be mapped with Rating color
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue="Rating color",palette=['white','red','orange','Yellow','Green','green'])
```

#This is how you add colors matching using palette
#Maximum number of rating is between 2.5 to 3.4 so we can handle the missing value using the mean of 2.5 and 3.4

```python
#Now we will use sns.countplot for categorical variables
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
